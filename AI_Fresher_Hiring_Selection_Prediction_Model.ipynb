```python
import pandas as pd
```

```python
#Read Dataset
df = pd.read_csv('https://github.com/YBIFoundation/ProjectDataSet/raw/main/Fresher%20Hiring%20Selection.csv')
```

```python
#Display Dataset Shape
df
```

```python
#Generate summary statistics.
df.describe()
```

```python
#Find percentage of selected candidates
Selected_percentage = (df['Selected'].mean()) *100
print('Selected Percentage',round(Selected_percentage,2),'%')
```

```python
df.columns
```

```python
#Define x and y
y=df['Selected']
x=df[['Age', 'CGPA', 'Aptitude_Score', 'Programming_Score',
       'SQL_Score', 'Communication_Score', 'Projects', 'Certifications',
       'Internship_Months', 'Hackathon_Participation', 'Attendance']]
```

```python
x.shape, y.shape
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=2)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#Select Model (Logistic Regression)
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
#Train Model
model.fit(x_train,y_train)
```

```python
#Prediction
y_pred = model.predict(x_test)
```

```python
#Evaluation
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
