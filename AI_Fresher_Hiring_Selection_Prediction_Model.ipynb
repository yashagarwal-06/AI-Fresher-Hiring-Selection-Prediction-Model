```python
#Boxplots for selected vs non-selected candidates.
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# CGPA
sns.boxplot(x = 'Selected', y = 'CGPA', data = df, ax = ax[0])
ax[0].set_title('CGPA vs Selection')

# Programming Score
sns.boxplot(x = 'Selected', y = 'Programming_Score', data = df, ax = ax[1])
ax[1].set_title('Programming Score vs Selection')

# SQL Score
sns.boxplot(x = 'Selected', y = 'SQL_Score', data = df, ax = ax[2])
ax[2].set_title('SQL Score vs Selection')

plt.tight_layout()
plt.show()
```

```python
#Define X(Features) and Y(Target)
Y = df['Selected']
X = df[['Age', 'CGPA', 'Aptitude_Score', 'Programming_Score',
       'SQL_Score', 'Communication_Score', 'Projects', 'Certifications',
       'Internship_Months', 'Hackathon_Participation', 'Attendance']]
```

```python
X.shape, Y.shape
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 2)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
#Select Model (Logistic Regression)
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
#Train Model
model.fit(X_train, Y_train)
```

```python
#Prediction
Y_pred = model.predict(X_test)
```

```python
#Evaluation
from sklearn.metrics import classification_report
```

```python
report = classification_report(Y_test, Y_pred)
print(report)
```
